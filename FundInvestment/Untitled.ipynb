```python
def sum(a, rate, years):
    q = 1 + rate
    return a*q*(q**years-1)/rate

def sum_rate(rate, years):
    q = 1 + rate
    return q*(q**years - 1)/(rate*years)
```

```python
a = 10000
rate = 0.15
years =30

print('收益率%d%%'%(rate*100), '定投每年%d元'%a, '定投%d年'%years) 
print('总收益:', sum(a, rate, years))
print('翻倍数:', sum_rate(rate, years))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year = np.arange(1, 31)
```

```python
bonus_sum = [sum(a, rate, y) for y in year]
bonus_rate = [sum_rate(rate, y) for y in year]
```

```python
plt.plot(year, bonus_sum)
```

```python
plt.plot(year, bonus_rate)
```
