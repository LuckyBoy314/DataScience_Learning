```python
from matplotlib import pyplot as plt
```

```python
# 当pyplot.plot()只有一个列表作为参数时,默认是plot(y),这时候的x轴是列表的索引
y = [4,2,7,4,10]
plt.ylabel('grade')
plt.plot(y)
#plt.show()
# 注释掉plt.show(), 可以看出当前图形的类别,即matplotlib.lines.Line2,但绘图过程中并没有引入这个类
# 而是直接使用了plt,所以说pyplot是一个命令快捷方式
```

```python
# 同时控制x,y轴的时候, 是plot(x,y),注意x轴在前,y轴在后
x = [2,4,6,8,10]
y = [4,2,7,4,10]

# 控制坐标轴的范围,四个参数分别表示 [x0,xn,y0,yn]
plt.axis([0,12, -1, 11])

plt.ylabel('grade')

plt.plot(x,y) # plt在调用savefig或show方法之前的顺序似乎无关紧要

# 第一个参数是保存路径和文件名,默认是png格式,dpi控制图片像素质量
plt.savefig('test',dpi=200) 

plt.show()
```

```python
import numpy as np
a = np.arange(10)

# plt.plot(x,y,format_string,[x, y, format_string])
# "x,y,format_stirng"三个参数可以反复出现,从而绘制多条曲线
# 绘制多条曲线的时候,x不能省略,但是format_string可以省略
# 绘制一条曲线的时候,x和format_string可以同时省略
plt.plot(a,a*1.5,a,a*2.5,a,a*3.5,a,a*4.5,a,a*5.5)
plt.show()
```

## format_string包含三种类型:
       'cms' 三个字母依次表示颜色/标记样式/线条样式
        cms顺序无关紧要
### 1. 颜色
####   用小写字母字符串表示
*  'r' 红
*  'g' 绿
*  'b' 蓝
*  'y' 黄
*  'w' 白
*  'k' 黑
*  'c' 青绿色 cyan
*  'm' 洋红色 magenta

#### 用RGB数字表示
    '#008034' 以十六进制表示三个分量
#### 灰度值用数字字符串表示
    '0.5' 

### 2. 样式(style)
*   '-' 实线
*   '--' 破折线
*   '-.' 点划线
*   ':' 虚线
*   ""  不写或者空字符串表示无线条

### 3. 标记(marker)

```python
plt.plot(a,a*1.5,'g-o',a,a*2.5, 'b:', a,a*3.5,a,a*4.5,a,a*5.5)
plt.show()
```

```python
import numpy as np
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

a = np.arange(0.0, 5.0, 0.02)
```

```python
plt.subplot(211)
plt.plot(a, f(a))
plt.subplot(2,1,2)
plt.plot(a, np.cos(2*np.pi*a),'r--')
plt.show()
```
